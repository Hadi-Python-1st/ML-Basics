Lenear regression is the relationship between two things such as an ice-cream and happiness in which we have to find out that how much ice-cream people eat and how much happy they are lenear regression is to draw a straight line between these two. The formule of lenear regression is **y = mx + b** where **y** is what you're trying to predict (happiness), **x** is what you know (ice cream), **m** is the slope (how much y changes when x changes), and **b** is the intercept (where the line starts).

# **2. Machine Learning(M.L)**

##**What is Machine Learning**

Machine learning is like teaching computers to learn from experience. Instead of programming them with specific instructions for every task, we give them lots of data and let them figure out patterns and make decisions on their own. It's kind of like how humans learn by seeing and doing things repeatedly.


##**How does it work's**

For example, if you show a computer many pictures of cats and dogs, it can learn to recognize which is which by finding patterns in the images. Over time, the computer gets better at identifying cats and dogs, even with new pictures it hasn't seen before.


##**Where it is used**

Machine learning is used in many areas, like recommending movies, recognizing speech, and even self-driving cars. It helps make predictions, find hidden insights, and automate decision-making processes based on data.


# **3. What are the types of Machine Learning**

**There are two types of Machine Learning**

##**Supervised Machine Learning**

Supervised machine learning is a type of machine learning where we teach the computer using examples that are labeled with the correct answers. Think of it like a teacher guiding a student.



##**Unsupervised Machine Learning**

Unsupervised machine learning is a type of machine learning where the computer is given data without any labels or correct answers. The goal is for the computer to find patterns and structure in the data on its own. It's like exploring without a map.

# **4. Supervised V.S Unsupervised**

##**Supervised Machine Learning**
1.   **Labled Data:** Uses labeled data, which means the training dataset includes input-output pairs. Each input has a corresponding correct output (label).
2.   **Training Goal:** The goal is to learn a mapping from inputs to outputs, allowing the model to predict the output for new, unseen inputs.

3.  **Example:** Email spam detection, where emails are labeled as "spam" or "not spam."

4.  **Applications:** Classification (e.g., identifying if an email is spam), regression (e.g., predicting house prices).

##**Unsupervised Machine Learning**

1. **Unlabled Data:** Uses unlabeled data, meaning the training dataset includes only the input data without any corresponding output labels.

2. **Training Goal:** The goal is to find hidden patterns, structures, or relationships within the input data.

3. **Example:**  Customer segmentation, where the goal is to group customers based on purchasing behavior without predefined categories.

4. **Applications:** Clustering (e.g., grouping similar items together), dimensionality reduction (e.g., reducing the number of features in a dataset).



# **5. What Does Regression and Classification means and how are the different**

**Regression and classification are two fundamental types of supervised machine learning tasks. Here’s what they mean and how they differ:**

## **Regression**

1. **Defination:** Regression is used to predict a continuous numeric value based on input data. It’s about finding a relationship between variables and using this relationship to predict an outcome.

2. **Example:** Predicting house prices based on features like size, location, and number of bedrooms.

3. **Output:** The output is a continuous value, such as a price, temperature, or age.

4. **Algorithms:** Common regression algorithms include Linear Regression, Polynomial Regression, and Support Vector Regression (SVR).

## **Classification**

1. **Defination:** Classification is used to predict a discrete category or class label for given input data. It’s about categorizing input data into predefined classes.

2. **Example:** Identifying whether an email is spam or not spam, or classifying images of animals as either cats or dogs.

3. **Output:** The output is a discrete value, such as class labels like "spam" or "not spam," "cat" or "dog."

4. **Algorithms:** Common classification algorithms include Logistic Regression, Decision Trees, Support Vector Machines (SVM), and Neural Networks.



# **6. What is the difference between training set and testing set**

**In machine learning, the training set and testing set are two different portions of the dataset used for building and evaluating a model. Here’s a breakdown of what each one is and how they differ:**

## **Training Set**

**What it is :**
* The training set is the subset of your data that the model learns from.
* It’s used to train the model by adjusting its parameters to find patterns in the data.

**Purpose :**
* To **fit the model:** The model uses the training data to learn how to make predictions or classifications.

* To **optimize parameter**s: The model adjusts its internal settings (weights, coefficients, etc.) based on the training data to minimize errors.

**Example:**

* If you’re building a model to predict house prices, the training set consists of historical house sales data that includes features like size, location, and number of bedrooms, along with the actual sale prices.

## **Testing Set**

**What it is :**
* The testing set is a separate subset of your data used to evaluate how well the model performs.

* It is not used during the training process.

**Purpose :**
* To **evaluate performance**: The model is tested on new, unseen data to see how well it generalizes to real-world scenarios.

* To **validate results**: It provides an unbiased evaluation of the model’s effectiveness.

**Example:**

* For the same house price prediction model, the testing set would consist of a different set of houses with known prices that the model has never seen during training.

# **7. What is E.D.A**

**Exploratory Data Analysis (EDA)**is a crucial step in the data science process. It involves exploring and analyzing your data to understand its structure, uncover patterns, and identify anomalies before building a machine learning model. Here’s a detailed overview of EDA:

## **What is EDA**
Exploratory Data Analysis (EDA) is an approach to analyzing datasets to summarize their main characteristics, often using visual methods. The goal is to explore data to understand it better and prepare it for further analysis.

## **Key Objectives of EDA**

1 : **Understand Data Characteristics:**
* Gain insights into the data’s distribution, relationships, and patterns.

* Identify trends, correlations, and potential issues.


2 : **Identify Data Quality Issues:**
* Detect missing values, outliers, and inconsistencies.
* Check for errors or anomalies in the data.

3 : **Generate Hypotheses:**
* Formulate initial hypotheses or questions based on data observations.
* Explore potential relationships and test assumptions.

4 : **Prepare Data for Modeling:**
* Clean and preprocess the data to get it ready for building models.
* Decide on feature selection, engineering, and transformation methods.

# **8. Linear Regressions / Multiple Linear Regressions**

## **Linear Regression**
* **What It Is:** A method to predict a value using just one feature.
* **When to Use It:** When you have one thing you’re measuring and one thing you’re trying to
* **Example:** Predicting someone's weight based on their height.
* **How It Works:** Finds the best straight line that represents the relationship between the height (feature) and weight (prediction).

## **Multiple Linear Regression**
* **What It Is:** A method to predict a value using two or more features.
* **When to Use It:** When you have multiple things that might affect what you’re trying to predict.
* **Example:** Predicting house prices based on size, location, and number of bedrooms.
* **How It Works:** Finds the best plane or surface that represents the relationship between the features (size, location, bedrooms) and the price.

## **When to Use Each Method**
* **Use Linear Regression** if you only have one feature to predict the outcome.
* **Use Multiple Linear Regression** if you have more than one feature that might influence the outcome.

## **Simple Definitios**
* **Linear Regression:** Predicts a result using one factor.
* **Multiple Linear Regression:** Predicts a result using several factors.

## **Summary**
In short, **Linear Regression** uses one predictor to find a relationship, while **Multiple Linear Regression** uses multiple predictors to find a more complex relationship.

## **Data Preparation**

* You can see that your dataset has many columns with values as 'Yes' or 'No'.
* But in order to fit a regression line, we would need numerical values and not string. Hence, we need to convert them to 1s and 0s, where 1 is a 'Yes' and 0 is a 'No'.

```python
varlist =  ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']

# Defining the map function
def binary_map(x):
    return x.map({'yes': 1, "no": 0})

# Applying the function to the housing list
housing[varlist] = housing[varlist].apply(binary_map)
```

**Rescaling the Features**

As you saw in the demonstration for Simple Linear Regression, scaling doesn't impact your model. Here we can see that except for area, all the columns have small integer values. So it is extremely important to rescale the variables so that they have a comparable scale. If we don't have comparable scales, then some of the coefficients as obtained by fitting the regression model might be very large or very small as compared to the other coefficients. This might become very annoying at the time of model evaluation. So it is advised to use standardization or normalization so that the units of the coefficients obtained are all on the same scale. As you know, there are two common ways of rescaling:

1. Min-Max scaling
2. Standardisation (mean-0, sigma-1)

This time, we will use MinMax scaling.

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

* **Apply scaler() to all the columns except the 'yes-no' and 'dummy' variables**

```python
num_vars = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking','price']

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

Dividing into X and Y sets for the model building

```python
y_train = df_train.pop('price')
X_train = df_train
```

## **Model Buildingt**

This time, we will be using the **LinearRegression function from SciKit** Learn for its compatibility with RFE (which is a utility from sklearn)

**RFE**

**R**ecursive **F**eature **E**limination

* **Importing RFE and LinearRegression**

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

* **Running RFE with the output number of the variable equal to 10**

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
rfe = RFE(estimator=lm, n_features_to_select=6)             # running RFE, use keyword argument for n_features_to_select
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

## **Building model using statsmodel, for the detailed statistics**

Creating X_test dataframe with RFE selected variables

```python
X_train_rfe = X_train[col]
```

* **Adding a constant variable**

```python
import statsmodels.api as sm
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
lm = sm.OLS(y_train,X_train_rfe).fit()   # Running the linear model
```

* **Let's see the summary of our linear model**

```python
print(lm.summary())
```

* **Calculate the VIFs for the model**

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
X = X_train_rfe
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```
